```python
# Assuming 'overall_Grade' is your target variable
X = df.drop('overall_Grade', axis=1)
y = df['overall_Grade']
```

Removing imbalance nature using SMOTE

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
# Instantiate SMOTE with a sampling strategy (you can adjust this)
smote = SMOTE(sampling_strategy='auto', random_state=42)

# Apply SMOTE to create synthetic samples
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Print the class distribution before and after SMOTE
print("Class distribution before SMOTE:", Counter(y))
print("Class distribution after SMOTE:", Counter(y_resampled))
```

Comparision between machine learning models

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression


# Reduce the size of the training data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.9, random_state=42)


# Instantiate the Logistic Regression model
lr_model = LogisticRegression()

# Fit the model on the training data
lr_model.fit(X_train, y_train)

# Predict labels on the testing data
ylr_pred = lr_model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, ylr_pred)
precision = precision_score(y_test, ylr_pred, average='macro')
recall = recall_score(y_test, ylr_pred, average='macro')
f1 = f1_score(y_test, ylr_pred, average='macro')

# Print evaluation metrics for Logistic Regression
print("Logistic Regression Evaluation Metrics:")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# Display the confusion matrix for Logistic Regression
cm = confusion_matrix(y_test, ylr_pred)
print("Confusion Matrix for Logistic Regression:")
print(cm)
```
